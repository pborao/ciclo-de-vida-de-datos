```python
import requests
import csv
from bs4 import BeautifulSoup
URL = 'https://www.volcanodiscovery.com/earthquakes/archive/2020-oct.html'
page = requests.get(URL)
soup = BeautifulSoup(page.text, 'html.parser')

q4 = soup.find_all('tr', class_='q4')
with open('q4.csv', mode='w') as csv_file:
    nroFila=1
    for job_elem in q4: 
        reportes_sismo = job_elem.find('a', class_='sl2')
        reportes_sismo = str(reportes_sismo).replace('<a class="sl2" href="','')
        reporte_tipo = reportes_sismo[:10]
        reporte_id = reportes_sismo[23:30] 
        coordenadas_sismo = job_elem.find_all('a', href = True, title = True, onclick = True, style = False) 
        coordenadas_sismo = str(coordenadas_sismo).replace(' - Show on map">Map</a>]','')
        coordenadas_sismo = coordenadas_sismo.split('title="')[1]
        names = []
        for celda in job_elem.find_all('td'):
            name=celda.text
            names.append(name)
        nroFila=nroFila+1
        names.append(reporte_tipo)
        names.append(reporte_id)
        names.append(coordenadas_sismo)
        writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerow(names)
        
q6 = soup.find_all('tr', class_='q6')
with open('q6.csv', mode='w') as csv_file:
    nroFila=1
    for job_elem in q6: 
        reportes_sismo = job_elem.find('a', class_='sl2')
        reportes_sismo = str(reportes_sismo).replace('<a class="sl2" href="','')
        reporte_tipo = reportes_sismo[:10]
        reporte_id = reportes_sismo[23:30] 
        coordenadas_sismo = job_elem.find_all('a', href = True, title = True, onclick = True, style = False) 
        coordenadas_sismo = str(coordenadas_sismo).replace(' - Show on map">Map</a>]','')
        coordenadas_sismo = coordenadas_sismo.split('title="')[1]
        names = []
        for celda in job_elem.find_all('td'):
            name=celda.text
            names.append(name)
        nroFila=nroFila+1
        names.append(reporte_tipo)
        names.append(reporte_id)
        names.append(coordenadas_sismo)
        writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerow(names)
        
q5 = soup.find_all('tr', class_='q5')
with open('q5.csv', mode='w') as csv_file:
    nroFila=1
    for job_elem in q5: 
        reportes_sismo = job_elem.find('a', class_='sl2')
        reportes_sismo = str(reportes_sismo).replace('<a class="sl2" href="','')
        reporte_tipo = reportes_sismo[:10]
        reporte_id = reportes_sismo[23:30] 
        coordenadas_sismo = job_elem.find_all('a', href = True, title = True, onclick = True, style = False) 
        coordenadas_sismo = str(coordenadas_sismo).replace(' - Show on map">Map</a>]','')
        coordenadas_sismo = coordenadas_sismo.split('title="')[1]
        names = []
        for celda in job_elem.find_all('td'):
            name=celda.text
            names.append(name)
        nroFila=nroFila+1
        names.append(reporte_tipo)
        names.append(reporte_id)
        names.append(coordenadas_sismo)
        writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerow(names)
        
q7 = soup.find_all('tr', class_='q7')
with open('q7.csv', mode='w') as csv_file:
    nroFila=1
    for job_elem in q7: 
        reportes_sismo = job_elem.find('a', class_='sl2')
        reportes_sismo = str(reportes_sismo).replace('<a class="sl2" href="','')
        reporte_tipo = reportes_sismo[:10]
        reporte_id = reportes_sismo[23:30] 
        coordenadas_sismo = job_elem.find_all('a', href = True, title = True, onclick = True, style = False) 
        coordenadas_sismo = str(coordenadas_sismo).replace(' - Show on map">Map</a>]','')
        coordenadas_sismo = coordenadas_sismo.split('title="')[1]
        names = []
        for celda in job_elem.find_all('td'):
            name=celda.text
            names.append(name)
        nroFila=nroFila+1
        names.append(reporte_tipo)
        names.append(reporte_id)
        names.append(coordenadas_sismo)
        writer = csv.writer(csv_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerow(names)
```
